```python
from sklearn.datasets import load_breast_cancer
import numpy as np
from sklearn.decomposition import KernelPCA
from mpl_toolkits.mplot3d import Axes3D

# Load the breast cancer dataset
data = load_breast_cancer()
X = data.data

# Define a custom cubic polynomial kernel
def cubic_polynomial_kernel(X1, X2):
    return (np.dot(X1, X2.T) + 1) ** 3

# Construct the kernel matrix
kernel_matrix = cubic_polynomial_kernel(X, X)

print(kernel_matrix)
import matplotlib.pyplot as plt

# Apply Kernel PCA
kpca = KernelPCA(n_components=3, kernel='precomputed')
X_kpca = kpca.fit_transform(kernel_matrix)

# Visualize the transfo rmed data in 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_kpca[:, 0], X_kpca[:, 1], X_kpca[:, 2], c=data.target, cmap='viridis')

# Add legend
legend1 = ax.legend(*scatter.legend_elements(), title="Classes")
ax.add_artist(legend1)

plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Load the iris dataset
iris = load_iris()
X_iris = iris.data
y_iris = iris.target

# Perform K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X_iris)

# Plot the clusters
plt.figure(figsize=(10, 6))
colors = ['r', 'g', 'b']
markers = ['o', 's', 'D']

for i in range(3):
    plt.scatter(X_iris[y_kmeans == i, 0], X_iris[y_kmeans == i, 1], 
                color=colors[i], marker=markers[i], label=f'Cluster {i+1}')

plt.title('K-Means Clustering of Iris Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

# Compute and show the confusion matrix
conf_matrix = confusion_matrix(y_iris, y_kmeans)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
